# Linear Regression Model Representation

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

### Training Data Initialization
In this section, the training data for a machine learning model is initialized. The input features are stored in the array `x_train_data`, and the corresponding output labels are stored in the array `y_train_labels`.

```python
# Initialize training data for a machine learning model
x_train_data = np.array([1.0, 2.0])
y_train_labels = np.array([300.0, 500.0])
```

### Display Training Data
Here, the training data, consisting of input features (`x_train_data`) and corresponding output labels (`y_train_labels`), is printed for visualization.

```python
# Print the training data
print(x_train_data)
print(y_train_labels)
```

### Number of Training Examples
The code snippet calculates and prints the number of training examples by querying the shape of the `x_train_data` array.

```python
# Get the number of training examples
num_training_examples = x_train_data.shape[0]
num_training_examples
```

### Visualization of Training Data
The code generates a scatter plot of the training data, illustrating the relationship between input features and output labels. The red 'x' markers represent the data points. The plot is titled "Scatter Plot of Training Data" with labeled axes.

```python
# Scatter plot of training data
plt.scatter(x_train_data, y_train_labels, marker="x", c="r")
plt.title("Scatter Plot of Training Data")
plt.xlabel("Input Features")
plt.ylabel("Output Labels")
plt.show()
```

### Model Parameters Initialization
In this section, the weight (`weight_parameter`) and bias (`bias_parameter`) parameters for a machine learning model are initialized with the specified values.

```python
# Initialize weight and bias parameters
weight_parameter = 200
bias_parameter = 100
```

### Model Output Computation Function
The function `compute_model_output` calculates the output of a linear regression model given input features, weight, and bias. The result is stored in the array `model_output`.

<p align="center">
    <img src="0_lin_rep.png" alt="Linear Regression Cost" />
</p>

```python
def compute_model_output(input_features, weight, bias):
    """
    Compute the output of a linear regression model.

    Parameters:
    - input_features: Input features as a NumPy array.
    - weight: Weight parameter of the linear regression model.
    - bias: Bias parameter of the linear regression model.

    Returns:
    - model_output: Output of the linear regression model.
    """
    m = input_features.shape[0]
    model_output = np.zeros(m)

    # Calculate the model output for each input feature
    for i in range(m):
        model_output[i] = weight * input_features[i] + bias

    return model_output
```

### Compute Model Output
Here, `temp_model_output` is calculated using the `compute_model_output` function. It represents the model output for the specified weight (`w`) and bias (`b`) applied to the training data (`x_train`).

```python
# Compute the model output for the given training data
temp_model_output = compute_model_output(input_features=x_train_data, weight=weight_parameter, bias=bias_parameter)
temp_model_output
```

### Comparison of Predicted and Actual Values
The code generates a plot comparing the model's predictions (`Our Prediction`, shown in blue) with the actual values (`Actual Values`, red 'x' markers). The plot is titled "Comparison of Predicted and Actual Values," and it visualizes the performance of the model on the training data.

```python
# Plotting the model prediction and actual values
plt.plot(x_train_data, temp_model_output, c="b", label="Our Prediction")
plt.scatter(x_train_data, y_train_labels, marker="x", c="r", label="Actual Values")
plt.title("Comparison of Predicted and Actual Values")
plt.xlabel("Input Features")
plt.ylabel("Output Labels")
plt.legend()
plt.show()
```
